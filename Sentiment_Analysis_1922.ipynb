```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
def remove_stopwords(tokens):
    stop_words =set(stopwords.words('english'))
    return [word for word in tokens if word not in stop_words]
```

```python
df['Clean'] = df['Clean'].apply(remove_stopwords)
```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
df['Clean'] = df['Clean'].apply(lambda tokens: ' '.join(tokens))
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(df['Clean'])
tfidf_vectors = tfidf_vectorizer.transform(df['Clean'])

```

```python
import numpy as np

# Convert a subset of TF-IDF vectors (e.g., first 100 rows) to a DataFrame
subset_size = 100  # You can adjust this to your desired subset size
subset_tfidf_df = pd.DataFrame(np.array(tfidf_vectors[:subset_size].toarray()), columns=tfidf_vectorizer.get_feature_names_out())

# Now you can inspect the subset_tfidf_df DataFrame
```

```python
import matplotlib.pyplot as plt

# Display original text and a subset of TF-IDF vectors as a bar chart for the first few rows
subset_size = 5
top_n_features = 10  # Choose the top N features to display

for i in range(subset_size):
    # Get the TF-IDF vector and corresponding feature names for the i-th row
    tfidf_vector = tfidf_vectors[i].toarray()[0]
    feature_names = tfidf_vectorizer.get_feature_names_out()

    # Sort the features by TF-IDF score and select the top N features
    top_features_idx = tfidf_vector.argsort()[-top_n_features:][::-1]
    top_features = [feature_names[idx] for idx in top_features_idx]
    top_tfidf_values = [tfidf_vector[idx] for idx in top_features_idx]

    # Create a bar chart for the top N features
    plt.figure(figsize=(8, 4))
    plt.barh(top_features, top_tfidf_values, color='skyblue')
    plt.xlabel('TF-IDF Score')
    plt.title(f'Top {top_n_features} TF-IDF Features for Row {i + 1}')
    plt.gca().invert_yaxis()  # Invert the y-axis for better readability
    plt.show()
    print("\n---\n")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize and train the Naive Bayes (Multinomial) classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)

# Make predictions on the test data using Naive Bayes
y_pred_nb = nb_classifier.predict(X_test)

# Calculate accuracy for Naive Bayes
accuracy_nb = accuracy_score(y_test, y_pred_nb)
# Calculate accuracy for Naive Bayes
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print(f"Naive Bayes Accuracy: {accuracy_nb * 100:.2f}%")

# Print the classification report for Naive Bayes
print("Classification Report for Naive Bayes:")
print(classification_report(y_test, y_pred_nb))

# Calculate the confusion matrix for Naive Bayes
conf_matrix_nb = confusion_matrix(y_test, y_pred_nb)

# Plot the confusion matrix as a heatmap for Naive Bayes
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nb, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for Naive Bayes')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get the log probabilities of features from the trained Naive Bayes (Multinomial) classifier
log_probabilities = nb_classifier.feature_log_prob_

# Calculate the absolute log probabilities
abs_log_probabilities = np.abs(log_probabilities)

# Select the top N features based on their absolute log probabilities (adjust N as needed)
top_n_features = min(5, abs_log_probabilities.shape[1])  # Choose the top 5 features
top_features_indices = abs_log_probabilities.argsort()[:, -top_n_features:]

# Get the corresponding feature names from the TF-IDF vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Initialize empty lists to store top features and coefficients
top_features = []
top_coefs = []

# Extract the top features and coefficients for each class using a loop
for i in range(log_probabilities.shape[0]):
    class_top_features = [feature_names[idx] for idx in top_features_indices[i]]
    class_top_coefs = [log_probabilities[i][idx] for idx in top_features_indices[i]]
    top_features.extend(class_top_features)
    top_coefs.extend(class_top_coefs)

# Create a bar chart for the top features and coefficients for Naive Bayes
plt.figure(figsize=(8, 4))
plt.barh(top_features, top_coefs, color='lightcoral')
plt.xlabel('Log Probability (Absolute Value)')
plt.title(f'Top {top_n_features} Features and Log Probabilities (Naive Bayes)')
plt.gca().invert_yaxis()

plt.show()
```
